```python
# Occurances
bucket = []
counts = df.groupby(['Total', 'Recap']).size()
bin_counts = counts[(counts > 1) & (counts < 3)]
for values in bin_counts.index:
    for idx in df.index[(df['Total'] == values[0]) & (df['Recap'] == values[1])]:
        bucket.append((idx, 0))
counts = counts[counts > 2]
for values in counts.index:
    for idx in df.index[(df['Total'] == values[0]) & (df['Recap'] == values[1])]:
        bucket.append((idx, 1))
```

```python
prizes = 4
bucket = []
counts = df.groupby(['Total', 'Recap']).size()
counts = counts[counts > 1]
counts.sort_index(ascending=False, inplace=True)
idx = 0
for entry in counts.index:
    logic = (df['Total'] == entry[0]) & (df['Recap'] == entry[1])
    candidates = df.index[logic]
    mode = 1 if logic[logic == True].size > 2 else 0
    if df.index.get_loc(candidates[0]) >= prizes:
        break
    bucket.extend([(candidate, mode) for candidate in candidates])
offset = df.index.get_loc(bucket[len(bucket) - 1][0]) + 1
df1 = df[:offset]
```

```python
# Extract groups of same occurances
# decor_occurance
prizes = 4
groups = []
counts = df.groupby(['Total', 'Recap']).size()
counts = counts[counts > 1]
counts.sort_index(ascending=False, inplace=True)
idx = 0
for entry in counts.index:
    logic = (df['Total'] == entry[0]) & (df['Recap'] == entry[1])
    candidates = df.index[logic]
    mode = 1 if logic[logic == True].size > 2 else 0
    if df.index.get_loc(candidates[0]) >= prizes:
        break
    groups.append(candidates.tolist())
#offset = df.index.get_loc(bucket[len(bucket) - 1][0]) + 1
#df1 = df[:offset]
groups
```
